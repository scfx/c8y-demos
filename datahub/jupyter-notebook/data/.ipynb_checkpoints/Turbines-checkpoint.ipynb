```python
sql = '''SELECT m.id, m.source as device, m.creationTime, m."Power.Actual.value", m."Wind.Speed.value", d.name as type
FROM t1989711161DataLake.rwe.t1989711161."measurements_1_minute_data" m
JOIN t1989711161Space."inventory_latest" d on m.source = d.id
WHERE m.creationTime >= DATE_SUB(CURRENT_TIMESTAMP, 2);'''
df = pd.read_sql(sql,cnxn)
df.describe()
```

## Power Curve

```python
turbine_types = df['type'].unique()
df = df[df['Power.Actual.value'] != 0]
print(turbine_types)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

scatter = ax.scatter(df['Wind.Speed.value'], df['Power.Actual.value'], c=df['type'].astype('category').cat.codes, cmap='viridis', s=30, alpha=0.8)
ax.set_xlabel('Wind [m/s]')
ax.set_ylabel('Power [kW]')
ax.set_title('Power Curve per Turbine Type')
handles, labels = scatter.legend_elements()
legend = plt.legend(handles, df['type'].unique(), title='Turbine', loc='upper left')
legend.get_title().set_fontsize('12')
plt.show()
```
